```python

import os
import cv2
import numpy as np
```

```python
def read_rawb_NirRGB(file_name):
    SMALL_IMAGE_SHAPE = (3454, 3650, 4)
    with open(file_name, 'rb') as file:
        img = file.read()
        
    img = np.frombuffer(img, dtype=np.float64)

    img = img.reshape(SMALL_IMAGE_SHAPE)

    return img.astype(np.uint8)

def read_rawb_RGB(file_name):
    img = read_rawb_NirRGB(file_name)
    
    return img[:, :, 1:4]  # Delete Nir channel
```

```python


path_to_masks = "/media/pablo.canosa/4373-30BA/Gaofen_Small_8bit/train/png_masks/"

# Iterate over the image names
image_names = []
for image_name in os.listdir(path_to_masks):
    # Read the image
    image = cv2.imread(f'{path_to_masks}{image_name}', cv2.IMREAD_UNCHANGED)

    # Check if the mask contains snow, square or stadium
    contains_snow = np.any(image == 16)
    contains_square = np.any(image == 20)
    contains_stadium = np.any(image == 19)

    # If the mask contains snow, square or stadium, print the image name and the classes
    if np.any(image == 16) :
        image_names.append(image_name)
        print(f'{image_name} contains snow')
```
