# Tamaño del marco

```python
image = cv2.imread(f'{GID_24_dataset_path_image}{filename}.tif')

height, width, _ = image.shape

middle_top = (width // 2, 0)
middle_bottom = (width // 2, height - 1)
middle_left = (0, height // 2)
middle_right = (width - 1, height // 2)

def count_pixels(image, start_point, step):
    count = 0
    current_color = image[start_point[1], start_point[0]]
    x, y = start_point
    while 0 <= x < width and 0 <= y < height and np.all(image[y, x] == current_color):
        count += 1
        x += step[0]
        y += step[1]
    return count

counts = {
    'top': count_pixels(image, middle_top, (0, 1)),
    'bottom': count_pixels(image, middle_bottom, (0, -1)),
    'left': count_pixels(image, middle_left, (1, 0)),
    'right': count_pixels(image, middle_right, (-1, 0))
}

print("Counts of border pixels:", counts)
```

Teniendo el cuenta en cambio en shape tiene sentido (6908, 7300, 4) vs (6800, 7200, 3)
