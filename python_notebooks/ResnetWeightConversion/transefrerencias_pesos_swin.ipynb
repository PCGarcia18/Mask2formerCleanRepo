```python
import pickle
from safetensors import safe_open

# Function to load safetensors file and save as pkl
def safetensors_to_pkl(input_file, output_file):
    # Open SafeTensors file
    tensors = {}
    with safe_open(input_file, framework="pt") as f:  # or 'tf' for TensorFlow
        print(len(f.keys()))
        for key in f.keys():
            tensors[key] = f.get_tensor(key)
    
    # Print keys and shapes
    print(len)
    for key, tensor in tensors.items():
        print(key, tensor.shape)

    # Dump to a .pkl file
    with open(output_file, 'wb') as f:
        pickle.dump(tensors, f)
    print(f"Saved to {output_file}")

# Example usage
input_file = '/home/pablo.canosa/wip/Mask2Former/backbone_weights/R50TESTS/bigEarthResnet.safetensors'
output_file = '/home/pablo.canosa/wip/Mask2Former/backbone_weights/R50TESTS/resnet_big_earth.pkl'

safetensors_to_pkl(input_file, output_file)
```
